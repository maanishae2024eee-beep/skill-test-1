```python
Question 1 (B) : Solve a System of Linear Equations
Problem Statement:
Write a Python program to solve a system of linear equations represented by:
Ax=b
Where:
● A is a 2x2 matrix (list of lists or NumPy array),
● b is a 2x1 vector,
● The program should compute and return the solution vector x.
```

```python
import numpy as np

A = np.array([[2,3],
              [1,2]])

b = np.array([5,3])

x = np.linalg.solve(A,b)

print("Solution x =", x)
```
