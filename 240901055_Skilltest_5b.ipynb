```python
Question5 (B): Analyze Healthcare Data Using Python and Pandas
Problem Statement:
You are provided with a CSV file named healthcare_data.csv that contains patient
information and their medical history. Write a Python script using Pandas to perform the
following tasks:
1. Load the dataset and display the column names along with their data types.
2. Calculate and display the average age of the patients.
3. Analyze the distribution of diagnoses to understand the most common medical
conditions in the dataset.
You may use Matplotlib or Seaborn to visualize the diagnosis distribution if desired.
```

```python
import numpy as np
import pandas as pd
health_df = pd.read_csv("healthcare_data.csv")
print("Column Names and Data Types:")
print(health_df.dtypes)
health_df = health_df.dropna()
average_age = np.mean(health_df['Age'])
print("\nAverage Age of Patients:", average_age)
diagnosis_distribution = health_df['Diagnosis'].value_counts()
print("\nDiagnosis Distribution:")
print(diagnosis_distribution)
```
