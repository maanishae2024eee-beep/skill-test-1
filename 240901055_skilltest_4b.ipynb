```python
Question 4(B): Perform Basic Data Analysis on Product Inventory Using
Pandas
Problem Statement:
You are given a dictionary containing information about products in an inventory, including
their names, prices, and quantities available:
data = {'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],'Price': [1200, 800,
300, 150, 50],'Quantity': [10, 20, 15, 25, 50]} and do any five basic operations and plot.
```

```python
import pandas as pd

data = {
'Product':['Laptop','Smartphone','Tablet','Monitor','Keyboard'],
'Price':[1200,800,300,150,50],
'Quantity':[10,20,15,25,50]
}

df = pd.DataFrame(data)

print("\nOriginal DataFrame:\n")
print(df)

# Mean
mean_values = pd.DataFrame(df.mean(numeric_only=True), columns=['Mean'])
print("\nMean values:\n")
print(mean_values)

# Sum
sum_values = pd.DataFrame(df.sum(numeric_only=True), columns=['Total'])
print("\nTotal values:\n")
print(sum_values)

# Maximum
max_values = pd.DataFrame(df.max(numeric_only=True), columns=['Maximum'])
print("\nMaximum values:\n")
print(max_values)

# Minimum
min_values = pd.DataFrame(df.min(numeric_only=True), columns=['Minimum'])
print("\nMinimum values:\n")
print(min_values)
```
