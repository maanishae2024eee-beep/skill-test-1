```python
Question 2(A) : Analyze Daily Step Count Using NumPy
Problem Statement:
Write a Python program using NumPy to analyze your daily step count data collected over
the past month. The program should:
● Compute key statistics such as total steps, average steps per day, and
maximum/minimum step counts.
● Identify days with high activity (e.g., days with steps above a certain threshold).
● Analyze trends over time, such as increases or decreases in activity, and optionally
visualize them.
```

```python
import numpy as np

steps = np.array([5000,6200,7000,8000,9000,7500,6200,
                  5400,10000,8500,9200,8800,7600,6400,
                  7200,8100,9300,8700,6900,7100])

print("Total steps:", steps.sum())
print("Average steps:", steps.mean())
print("Maximum steps:", steps.max())
print("Minimum steps:", steps.min())

# High activity days
high = steps[steps > 8000]
print("High activity days:", high)
```
