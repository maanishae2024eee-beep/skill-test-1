```python
Question 4 (A) : Analyze EEG Data Using Pandas
Problem Statement:
You have collected EEG data from multiple participants while they performed a cognitive
task. Each dataset includes time-series measurements of brainwave activity across various
electrodes. Write a Python program using Pandas to:
● Load and organize the EEG data.
● Perform basic statistical analysis (mean, variance, peak values) for each participant
and electrode.
● Identify patterns or trends in brainwave activity during the task.
```

```python
import pandas as pd

data = {
    'Participant':['P1','P2','P3','P4'],
    'Electrode1':[12,15,14,16],
    'Electrode2':[20,18,19,21]
}

df = pd.DataFrame(data)

print("\nOriginal DataFrame:\n")
print(df)

# Mean
mean_values = pd.DataFrame(df.mean(numeric_only=True), columns=['Mean'])
print("\nMean values:\n")
print(mean_values)

# Variance
variance_values = pd.DataFrame(df.var(numeric_only=True), columns=['Variance'])
print("\nVariance values:\n")
print(variance_values)

# Maximum
max_values = pd.DataFrame(df.max(numeric_only=True), columns=['Maximum'])
print("\nMaximum values:\n")
print(max_values)
```
