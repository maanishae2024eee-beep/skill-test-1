```python
Question 2(B) : Analyze Student Grades Using NumPy
Problem Statement:
Write a Python program using NumPy to analyze the grades of 30 students across three
subjects: Math, Science, and English. Your program should:
● Calculate overall and subject-wise statistics such as average, highest, and lowest
grades.
● Identify top-performing students based on total or average scores.
● Optionally visualize the grade distributions for each subject using histograms or bar
charts.
```

```python
import numpy as np

grades = np.random.randint(50, 100, size=(30,3))   # 30 students, 3 subjects, grades between 50 and 100

print("Grades Matrix:\n", grades)

print("Subject Average:", grades.mean(axis=0))
print("Highest Marks:", grades.max())
print("Lowest Marks:", grades.min())

total = grades.sum(axis=1)
top_student = total.argmax()

print("Top Student Index:", top_student)
```
