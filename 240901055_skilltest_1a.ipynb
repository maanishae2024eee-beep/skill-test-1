```python
Question 1(A) : Analyze Student Exam Scores with Python
Problem Statement:
You are provided with the exam scores of 20 students in a class. Write a Python program to
analyze these scores. Your analysis should include:
1. Overall class performance – Calculate average, highest, and lowest scores.
2. Pass/fail classification – Identify which students passed the exam (assume pass mark
is 50).
3. Grading on a curve – Assign letter grades (A, B, C, D, F) based on curved scoring or
predefined thresholds.
```

```python
# Student scores
scores = [65, 72, 45, 80, 55, 60, 49, 90, 33, 77,
          58, 69, 74, 82, 47, 53, 61, 88, 92, 40]

# Average, highest, lowest
avg = sum(scores) / len(scores)
highest = max(scores)
lowest = min(scores)

print("Average Score:", avg)
print("Highest Score:", highest)
print("Lowest Score:", lowest)

# Pass / Fail
for s in scores:
    if s >= 50:
        print(s, "Pass")
    else:
        ("Fail")

# Grade classification
for s in scores:
    if s >= 85:
        grade = "A"
    elif s >= 70:
        grade = "B"
    elif s >= 60:
        grade = "C"
    elif s >= 50:
        grade = "D"
    else:
        grade = "F"

    print(s, "Grade:", grade)
```
